```python
from selenium import webdriver
from selenium.webdriver.common.by import By
import time
import pandas as pd
```

```python
driver = webdriver.Chrome()
url = r'https://www.technolife.ir/'
driver.maximize_window()
driver.get(url)
driver.find_element(By.XPATH, """/html/body/div[1]/div[3]/header/div/header/nav/ul/li[1]/div/span""").click() #mobile&tablet
driver.find_element(By.XPATH, """/html/body/div[1]/div[3]/header/div/header/nav/ul/li[1]/div/div/nav/ul/li[1]/div/span[2]""").click() #mobile
driver.find_element(By.XPATH, """/html/body/div[1]/div[3]/header/div/header/nav/ul/li[1]/div/div/nav/div/div[2]/ul[1]/li[3]/a""").click() #Xiaomi mobile

#scrol
avaliable = driver.find_element(By.CSS_SELECTOR, f"""#__next > div.w-full > main > div > div.mt-8.flex.gap-8 > div.h-full.w-full.flex-grow >
                                 div.sticky.flex.w-full.items-start.bg-white.py-6.-mt-6.z-20.-mr-1.descktop_top_stuck > div.flex-grow > div > div:nth-child(2) > p""").text
avaliable = avaliable.split(" کالا") # separate
avaliable = int(avaliable[0])
total_scroll = int(avaliable / 3)
scroll_to = 396
for scroll_t in range(total_scroll):
    scroll = driver.execute_script(f"window.scrollTo(0, {scroll_to})")
    scroll_to += 396
    time.sleep(1)
scroll = driver.execute_script("window.scrollTo(0, 0)")
```

```python
title = driver.find_elements(By.XPATH, """/html/body/div[1]/div[3]/main/div/div[1]/div[2]/div[2]/div/div/div[3]/a""")
price = driver.find_elements(By.XPATH, """/html/body/div[1]/div[3]/main/div/div[1]/div[2]/div[2]/div/div/div[4]/div""")

title_list = []
price_list = []

for i in title:
    j = i.text
    title_list.append(j)
print(title_list)
print(len(title_list))

for i in price:
    j = i.text
    price_list.append(j)
print(price_list)
print(len(price_list))
```

```python
#پیدا کردن محصولات با تخفیف

off_price = driver.find_elements(By.XPATH, """/html/body/div[1]/div[3]/main/div/div[1]/div[2]/div[2]/div/div/div[4]/div/div[2]/p""")
off_price_list = []
for i in off_price:
    j = i.text
    off_price_list.append(j)
print(off_price_list)
print(len(off_price_list))
```

```python
import re
price_list = [sub.replace('تومان', '') for sub in price_list]
off_price_list = [sub.replace('تومان', '') for sub in off_price_list]

print(price_list)
print(len(price_list))

print(off_price_list)
print(len(off_price_list))

def replace_items(first_list, second_list, target_character):
    for i in range(len(first_list)):
        if first_list[i].count(target_character) == 3:
            first_list[i] = second_list[0]
            second_list.pop(0)
    return first_list

target_character = "\n"
new_list = replace_items(price_list, off_price_list, target_character)
print(new_list)
print(len(new_list))
```

```python
# create dataframe
data = {'Title': title_list,
        'Price': new_list}
df = pd.DataFrame(data)
df['Price'] = df['Price'].replace({',': ''}, regex=True)
df
```

```python
df = df.drop(df[(df['Price'] == 'ناموجود')].index)     # delete items
df['Price'] = pd.to_numeric(df['Price'])      # change data type from object to int
df
```

```python
df.to_csv('xiaomi_technolife.csv')
```
