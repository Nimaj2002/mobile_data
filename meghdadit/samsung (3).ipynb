```python
import pandas as pd 
import re
from selenium import webdriver
from selenium.webdriver.common.by import By
from selenium.webdriver.support.ui import WebDriverWait
import time
```

```python

driver = webdriver.Chrome()
driver.maximize_window()
url="""https://meghdadit.com/productlist/mobile/?b=82"""
driver.get(url) 
```

```python

    
    


t_models=[]
t_prices=[]

i = 1
while i<=2:
    driver.get(driver.current_url)
    models = driver.find_elements(By.XPATH, """/html/body/form/div[3]/div/div[3]/div/div[2]/article/div[1]/ul/li/div/div[4]/div[1]/a""")
    prices = driver.find_elements(By.XPATH, """/html/body/form/div[3]/div/div[3]/div/div[2]/article/div[1]/ul/li/div/div[4]/div[2]/div/div/span""")

    
        
    for model in models :
        name= model.text
        t_models.append(name)
    for price in prices:
        Price=price.text
        t_prices.append(Price)
    
    n_prices = [re.sub(r'[,تومان]','',price) for price in t_prices]
    
    
    next_btn =driver.find_element(By.XPATH, """/html/body/form/div[3]/div/div[3]/div/div[2]/article/div[2]/div/a[7]""")
    next_btn.click()
    i+=1

print(t_models)
print(n_prices)



    

```

```python
data = {'model': t_models,'Price': n_prices}
df = pd.DataFrame(data) 
```

```python
df = df.drop(df[(df['Price'] == 'جد')].index)
df
```

```python
df.to_csv('samsung_phones')
```
